```python
home_data_for_saleDc = home_data_for_sale.dropna(axis=1) # axis = 0/1 delete row/columns with missing value
```

```python
home_data_for_saleDc.shape
```

```python
y = home_data_for_saleDc.SalePrice #SalePrice is target
home_data_for_saleDc_features = ['LotArea','YearRemodAdd','YearBuilt','BedroomAbvGr','GarageCars']
x = home_data_for_saleDc[home_data_for_saleDc_features] # [selected features data]
```

```python
from sklearn.model_selection import train_test_split # Spliting data into train and cross validation sets.

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3, random_state=1)


from sklearn.tree import DecisionTreeRegressor  #To select method.

model = DecisionTreeRegressor(random_state=1)

model.fit(x_train, y_train)     #To fit model.
```

```python
y_predict = model.predict(x_test)
# print(y_predict)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_predict) #To compare function.
```

```python
# Define percentage error as a function.
def mean_perc_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
# Define logarithm error as a function.
def mean_log_error(y_true, y_pred): 
    return np.mean(np.abs((np.log(y_true) - np.log(y_pred))))
```

```python
mean_perc_error(y_test, y_predict)

```

```python

mean_log_error(y_test, y_predict)
```
